```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def check_adult(self):
        if self.age >= 20:
            return "大人"
        else:
            return "未成年"

user1 = Human("A", 19)
user2 = Human("B", 26)
user3 = Human("C", 20)
user4 = Human("D", 21)

users = [user1, user2, user3, user4]

for user in users:
    print(f"{user.name}は{user.check_adult()}")
```
